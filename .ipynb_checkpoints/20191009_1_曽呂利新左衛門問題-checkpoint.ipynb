## 問題１ 100日目の米粒の数
100日目には合計で米粒をいくつもらっているのかを計算するコードを作成


```python
import matplotlib.pyplot as plt
%matplotlib inline

rice = 1
day = 100
today_rice = 0
today_rice_list = []
sum_rice_list = []

for i in range(day):
    today_rice = rice
    today_rice_list.append(today_rice)
    rice *= 2
    sum_rice_list.append(sum(today_rice_list))

print("その日にもらう米の数", today_rice_list)
print("その日までに累計でもらう米の数", sum_rice_list)

#グラフの表示
plt.title("rice number of the day")
plt.xlabel("day")
plt.ylabel("rice number")
plt.plot(today_rice_list, marker = ".") #マーカーいれる。
plt.show()

plt.title("total rice number")
plt.xlabel("day")
plt.ylabel("rice number")
plt.plot(sum_rice_list, marker = ".") #マーカーいれる。
plt.show()
```

## 問題２ 100日目以外の米粒の数
・日数を引数とした 関数 を作成する
・コースを受講する日数の間でいくつ米粒をもらえるか算出する


```python
import matplotlib.pyplot as plt
%matplotlib inline

def rice_number(days):
    """
    引数で与えられた日数を２のべき乗で計算する関数
    """
    return 2 ** days

def make_list(days):
    day_rice_number_list = []  #その日にもらう米の数
    total_rice_number = 0
    total_rice_number_list = []  #その日までの合計米数格納リスト

    for i in range(days):
        day_rice_number_list.append(rice_number(i))
        total_rice_number += day_rice_number_list[i]
        total_rice_number_list.append(total_rice_number)
        
    #グラフの表示　その日にもらう米数
    plt.title("rice number of the day")
    plt.xlabel("day")
    plt.ylabel("rice number")
    plt.plot(day_rice_number_list, marker = ".") #マーカーいれる。
    plt.show()
        
    #グラフの表示　累計のもらう米数
    plt.title("total rice number of the day")
    plt.xlabel("day")
    plt.ylabel("rice number")
    plt.plot(total_rice_number_list, marker = ".") #マーカーいれる。
    plt.show()
    
    return day_rice_number_list, total_rice_number_list


make_list(7)

```

```python
#コース受講する期間にもらえる米粒
#コース受講期間＝123日

course_day_rice, course_total_rice = make_list(123)
print("コース最終日にもらえる累計米数＝{}粒" .format(course_total_rice[-1]))
```

## 【問題３】何人が何日生活できる米か
累計米数で何人が何日生活できるか計算する関数を作成する。


なお、１人が1日で必要とする米数は、1日に必要とするカロリーを2000kcalとした場合、


1合を約500kcal、6,500粒と想定すると、4合＝26,000粒あれば良いと考えることとする。

参考：https://kw-note.com/trivia/kome-1gou/

```python
#生活できる米の計算

def life_req_rice(rice_num, people_num):
    day_req_rice = people_num * 26000  #1日に必要な米数
    day_long = rice_num / day_req_rice
    return day_long

life_req_rice(course_total_rice[-1], 12)
```
