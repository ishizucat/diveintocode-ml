### 【問題3】標準化クラスをスクラッチで作成
---

```python
class ScratchStandardScaler():
    """
    標準化のためのクラス

    Attributes
    ----------
    mean_ : 次の形のndarray, shape(n_features,)
        平均
    var_ : 次の形のndarray, shape(n_features,)
        分散
    """

    def fit(self, X):
        """
        標準化のために平均と標準偏差を計算する。

        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            学習データ
        """
        self.mean_ = np.array([])
        self.var_ = np.array([])
        self.std_ = np.array([])
        cols = X.shape[1]  #列数の取得
        raws = X.shape[0]  #行数の取得

        for i in range(cols):
            self.mean_ = np.append(self.mean_, X[:, i].sum() / X[:, i].shape)
        
        for i in range(cols):
            var = 0
            for j in range(raws):
                var += (X[j, i] - self.mean_[i])**2
            self.var_ = np.append(self.var_, var / raws)
            self.std_ = np.append(self.std_, np.sqrt(var / raws))

        pass

    def transform(self, X):
        """
        fitで求めた値を使い標準化を行う。

        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            特徴量

        Returns
        ----------
        X_scaled : 次の形のndarray, shape (n_samples, n_features)
            標準化された特緒量
        """
        cols = X.shape[1]  #列数の取得
        raws = X.shape[0]  #行数の取得
        X_scaled = np.array([])
        temp = np.array([])
        
        for j in range(raws):
            temp = np.append(temp, (X[j, 0] - self.mean_[0])/self.std_[0])
        X_scaled = temp.reshape(raws, 1)
        
        for i in range(1, cols):
            temp = np.array([])
            for k in range(raws):
                temp = np.append(temp, (X[k, i] - self.mean_[i])/self.std_[i])
            temp = temp.reshape(raws, 1)
            X_scaled = np.hstack((X_scaled, temp))

        
        pass
        return X_scaled
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris

data = load_iris()
X = data.data[:10]

scaler = StandardScaler()
scaler.fit(X)
print("平均 :", scaler.mean_)
print("分散 :", scaler.var_)
X_std = scaler.transform(X)
print(X_std)
```

```python

print(X)
scratch_scaler = ScratchStandardScaler()
scratch_scaler.fit(X)
print("平均 : {}".format(scratch_scaler.mean_))
print("分散 : {}".format(scratch_scaler.var_))
print("標準偏差 : {}".format(scratch_scaler.std_))
X_std = scratch_scaler.transform(X)
print(X_std)
```

【問題4】 四則演算を行うクラスの作成
---

```python
class FourOpeClass():
    """
    説明用の簡単なクラス

    Parameters
    ----------
    value : float or int
        初期値

    Attributes
    ----------
    value : float or int
        計算結果
    """
    def __init__(self, value):
        try:
            val = int(value)
            self.value = val
            print("初期値{}が設定されました".format(self.value))
        
        except ValueError:
           print("That's not an int!")
            
    def add(self, value2):
        """
        受け取った引数をself.valueに加える
        """
        self.value += value2
    
    def sub(self, value2):
        """
        受け取った引数をself.valueから引く
        """
        self.value -= value2

    def multi(self, value2):
        """
        受け取った引数をself.valueとかける
        """
        self.value *= value2
   
    def div(self, value2):
        """
        self.valueを受け取った引数で割る
        """
        self.value /= value2
        
example = FourOpeClass(10.5)
print("value : {}".format(example.value))
example.div(5)
print("value : {}".format(example.value))
```
